```python
import pandas as pd

# 첫 번째 행(헤더)을 제외하고 모든 열을 숫자로 변환
df_numeric = sejong.iloc[1:].apply(pd.to_numeric, errors='coerce')

# 평균 계산
column_averages = df_numeric.mean()

# 평균 출력
print(column_averages)
```

```python
column_averages=column_averages.T
```

```python
column_averages
```

```python
# Series를 DataFrame으로 변환하고 transpose 적용
df_transposed = pd.DataFrame(column_averages).transpose()
```
