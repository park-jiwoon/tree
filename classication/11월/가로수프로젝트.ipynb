```python
import numpy as np
import pandas as pd

df2 = df1.iloc[:,4:20]
```

```python
df2.columns = ["체리나무", "은행나무 ", "자쿠라나무 ", "단풍나무 "," 메이플 나무", "메타세콰이 나무", "작은 소나무","탈로우나무",
                                             "소나무", "한국소나무","침엽수", "헤녹카이프러스 나무", "아카시아나무",
                                             "매그놀리아나무", "참나무","기타"]
```

```python
df2 #  tCO2eq
```

```python
df3 = df2.T
```

```python
df3.isnull().sum()
```

```python
df3
```

```python
for col in df3.columns:
    # 비숫자 문자 제거 (예: 쉼표)
    df3[col] = df3[col].replace(',', '', regex=True)

    # 열을 숫자로 변환
    df3[col] = pd.to_numeric(df3[col], errors='coerce')

# 결과 확인
print(df3.info())
```

```python
# 열별 합계 계산
df3['Total_Sum'] = df3.sum(axis=1)
# 결과 확인
print(df3)
```

```python
df3
```

```python
# 열별 평균 추가
df3['Average'] = df3.mean(axis=1)

# 결과 확인
print(df3)
```
