```python
import pandas as pd
from geopy.geocoders import Nominatim

# 예시 데이터셋을 불러오기 (실제 경로로 대체해야 함)
# data = pd.read_csv('your_data_file.csv')

# 지오코더 초기화
geolocator = Nominatim(user_agent="geoapiExercises")

# 예시 데이터 프레임 생성 (실제 데이터로 대체해야 함)
data = pd.DataFrame({
    'Hospital': ['충남대학교병원', '대전성모병원'],
    'Address': ['대전광역시 중구 문화로 282', '대전광역시 중구 대흥로 64'],
    'Beds': [1316, 664]
})

# 주소를 위도와 경도로 변환하는 함수
def geocode(address):
    try:
        location = geolocator.geocode(address)
        return (location.latitude, location.longitude)
    except:
        return (None, None)

# 새로운 'Latitude'와 'Longitude' 열을 데이터 프레임에 추가
data[['Latitude', 'Longitude']] = data['Address'].apply(geocode).tolist()

print(data)
```

```python
import folium

# 데이터셋에 대한 가정: 'Latitude'와 'Longitude' 열이 포함되어 있음
# data = pd.read_csv('your_data_file.csv')  # 실제 데이터셋 불러오기

# 지도 생성, 중심점은 데이터의 첫 번째 위치로 설정
m = folium.Map(location=[data.iloc[0]['Latitude'], data.iloc[0]['Longitude']], zoom_start=12)

# 데이터셋의 각 위치에 대한 마커 추가
for idx, row in data.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=row['Hospital']).add_to(m)

# 지도를 HTML 파일로 저장
m.save('map.html')
```
